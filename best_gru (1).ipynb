```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import GRU, Dense, Dropout, BatchNormalization
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint
import glob
import os
from google.colab import drive

# Mount Google Drive
drive.mount('/content/drive')

# Set your data path - UPDATED PATH
DATA_PATH = "/content/drive/MyDrive/HUPA-UCM Diabetes Dataset (1)/HUPA-UCM Diabetes Dataset/Preprocessed"

# Check all files in the directory
import os
all_files = os.listdir(DATA_PATH)
print("All files in directory:")
for f in all_files:
    print(f" - {f}")

# Filter for CSV files
csv_files = [f for f in all_files if f.endswith('.csv')]
print(f"\nFound {len(csv_files)} CSV files:")
for f in csv_files:
    print(f" - {f}")

def load_all_patients_data(data_path, csv_files):
    patients_data = {}

    for filename in csv_files:
        try:
            filepath = os.path.join(data_path, filename)

            # Extract patient number from filename
            patient_num_str = filename.replace('HUPA', '').replace('P.csv', '').replace('.csv', '').lstrip('0')
            patient_num = int(patient_num_str) if patient_num_str else 1

            # Load with proper semicolon separation
            df = pd.read_csv(filepath, sep=';')

            # Convert time column to datetime
            df['time'] = pd.to_datetime(df['time'], errors='coerce')

            # Drop rows with invalid time
            df = df.dropna(subset=['time'])

            # Sort by time
            df = df.sort_values('time').reset_index(drop=True)

            # Convert numeric columns to appropriate types
            numeric_columns = ['glucose', 'calories', 'heart_rate', 'steps',
                             'basal_rate', 'bolus_volume_delivered', 'carb_input']

            for col in numeric_columns:
                if col in df.columns:
                    df[col] = pd.to_numeric(df[col], errors='coerce')

            patients_data[patient_num] = df

        except Exception as e:
            print(f"Error loading {filename}: {str(e)}")
            continue

    print(f"\nSuccessfully loaded {len(patients_data)} patients")
    return patients_data

# Load all patient data with proper semicolon separation
all_patients = load_all_patients_data(DATA_PATH, csv_files)

if all_patients:
    first_patient_id = list(all_patients.keys())[0]
    first_patient_data = all_patients[first_patient_id]

    print("First patient data shape:", first_patient_data.shape)
    print("\nColumns:", first_patient_data.columns.tolist())
    print("\nFirst few rows:")
    display(first_patient_data.head(10))

    # Data info
    print("\nData types:")
    print(first_patient_data.dtypes)

    # Check for missing values
    print("\nMissing values per column:")
    print(first_patient_data.isnull().sum())

    # Check time range and frequency
    print(f"\nTime range: {first_patient_data['time'].min()} to {first_patient_data['time'].max()}")
    time_diff = first_patient_data['time'].diff().mean()
    print(f"Average time difference: {time_diff}")

    # Basic statistics
    print("\nBasic statistics:")
    display(first_patient_data.describe())

else:
    print("No patients loaded. Let's debug the file structure...")

    # Debug: Check raw file content
    if csv_files:
        sample_file = os.path.join(DATA_PATH, csv_files[0])
        print(f"\nDebugging file: {csv_files[0]}")

        # Read raw file content
        with open(sample_file, 'r') as f:
            lines = f.readlines()

        print("First 3 lines of raw file:")
        for i, line in enumerate(lines[:3]):
            print(f"Line {i}: {line.strip()}")

if all_patients:
    print("Data Quality Summary:")

    for patient_id, data in list(all_patients.items())[:5]:  # Check first 5 patients
        print(f"\nPatient {patient_id}:")
        print(f"  Records: {len(data)}")
        print(f"  Time span: {data['time'].max() - data['time'].min()}")
        print(f"  Glucose range: {data['glucose'].min():.1f} - {data['glucose'].max():.1f}")
        print(f"  Missing values: {data.isnull().sum().sum()}")

        # Check if we have insulin delivery data
        if 'bolus_volume_delivered' in data.columns:
            bolus_records = data[data['bolus_volume_delivered'] > 0]
            print(f"  Bolus records: {len(bolus_records)}")

        if 'carb_input' in data.columns:
            carb_records = data[data['carb_input'] > 0]
            print(f"  Carb intake records: {len(carb_records)}")

def preprocess_patient_data(patient_df, target_glucose=110, insulin_sensitivity_factor=50):
    df = patient_df.copy()

    # Ensure numeric types
    numeric_columns = ['glucose', 'calories', 'heart_rate', 'steps',
                      'basal_rate', 'bolus_volume_delivered', 'carb_input']

    for col in numeric_columns:
        if col in df.columns:
            df[col] = pd.to_numeric(df[col], errors='coerce')

    # Feature engineering
    df['glucose_change'] = df['glucose'].diff()  # Rate of change
    df['glucose_rolling_mean'] = df['glucose'].rolling(window=3, min_periods=1).mean()
    df['calories_cumulative'] = df['calories'].rolling(window=6, min_periods=1).sum()
    df['steps_cumulative'] = df['steps'].rolling(window=6, min_periods=1).sum()

    # Time-based features
    df['hour'] = df['time'].dt.hour
    df['minute'] = df['time'].dt.minute
    df['sin_hour'] = np.sin(2 * np.pi * df['hour'] / 24)
    df['cos_hour'] = np.cos(2 * np.pi * df['hour'] / 24)

    # Calculate insulin need target (simplified correction dose)
    # Using standard correction formula: (Current Glucose - Target) / ISF
    df['insulin_need'] = np.maximum(0, (df['glucose'] - target_glucose)) / insulin_sensitivity_factor

    # Use future insulin need as target (shifted by 1 time step = 5 minutes)
    df['target_insulin_need'] = df['insulin_need'].shift(-1)

    # Drop NaN values created by shifting and rolling
    df = df.dropna()

    return df

def create_sequences_for_patient(patient_data, sequence_length=6):
    """
    Create sequences for a single patient
    """
    # Selected features for the model
    features = [
        'glucose', 'calories', 'heart_rate', 'steps',
        'basal_rate', 'bolus_volume_delivered', 'carb_input',
        'glucose_change', 'glucose_rolling_mean',
        'calories_cumulative', 'steps_cumulative',
        'sin_hour', 'cos_hour'
    ]

    # Only use features that exist in the dataframe
    available_features = [f for f in features if f in patient_data.columns]

    X, y = [], []

    for i in range(len(patient_data) - sequence_length):
        # Input sequence (last 30 minutes = 6 time steps * 5min)
        sequence_features = patient_data[available_features].iloc[i:i+sequence_length].values
        X.append(sequence_features)

        # Target: Insulin need in next period (after the sequence)
        target = patient_data['target_insulin_need'].iloc[i+sequence_length-1]
        y.append(target)

    return np.array(X), np.array(y)

def prepare_all_patients_data(all_patients_data, sequence_length=6):
    # Patient IDs
    patient_ids = list(all_patients_data.keys())
    np.random.seed(42)
    np.random.shuffle(patient_ids)

    # Split patients (20 train, 3 val, 2 test for 25 patients)
    train_patients = patient_ids[:20]
    val_patients = patient_ids[20:23]
    test_patients = patient_ids[23:]

    print(f"Train patients: {train_patients}")
    print(f"Val patients: {val_patients}")
    print(f"Test patients: {test_patients}")

    X_train, y_train = [], []
    X_val, y_val = [], []
    X_test, y_test = [], []

    # Process each patient
    for patient_id, patient_data in all_patients_data.items():
        # Preprocess patient data
        processed_data = preprocess_patient_data(patient_data)

        # Create sequences
        X_patient, y_patient = create_sequences_for_patient(processed_data, sequence_length)

        # Assign to appropriate split
        if patient_id in train_patients:
            X_train.append(X_patient)
            y_train.append(y_patient)
        elif patient_id in val_patients:
            X_val.append(X_patient)
            y_val.append(y_patient)
        else:  # test patients
            X_test.append(X_patient)
            y_test.append(y_patient)

    # Concatenate all sequences
    if X_train:
        X_train = np.concatenate(X_train, axis=0)
        y_train = np.concatenate(y_train, axis=0)
    if X_val:
        X_val = np.concatenate(X_val, axis=0)
        y_val = np.concatenate(y_val, axis=0)
    if X_test:
        X_test = np.concatenate(X_test, axis=0)
        y_test = np.concatenate(y_test, axis=0)

    print(f"\nFinal dataset shapes:")
    print(f"Training set: {X_train.shape} sequences")
    print(f"Validation set: {X_val.shape} sequences")
    print(f"Test set: {X_test.shape} sequences")

    return (X_train, y_train), (X_val, y_val), (X_test, y_test), (train_patients, val_patients, test_patients)

# Prepare the data
(X_train, y_train), (X_val, y_val), (X_test, y_test), patient_splits = prepare_all_patients_data(all_patients)

print(f"Data prepared!")
print(f"X_train shape: {X_train.shape}")
print(f"y_train shape: {y_train.shape}")

# Get the number of features
n_features = X_train.shape[2]
print(f"Number of features: {n_features}")

from tensorflow.keras.layers import LeakyReLU

def create_insulin_gru_deep(sequence_length=6, n_features=13):
    model = Sequential([
        GRU(128, return_sequences=True, input_shape=(sequence_length, n_features),
             kernel_initializer='he_normal', recurrent_dropout=0.2),
        BatchNormalization(),
        Dropout(0.3),

        GRU(64, return_sequences=True, kernel_initializer='he_normal', recurrent_dropout=0.2),
        BatchNormalization(),
        Dropout(0.3),

        GRU(32, return_sequences=False, kernel_initializer='he_normal', recurrent_dropout=0.1),
        Dropout(0.2),

        Dense(64, kernel_initializer='he_normal'),
        LeakyReLU(alpha=0.1),
        BatchNormalization(),
        Dropout(0.2),

        Dense(32, kernel_initializer='he_normal'),
        LeakyReLU(alpha=0.1),
        Dropout(0.1),

        Dense(16, activation='relu', kernel_initializer='he_normal'),
        Dense(1, activation='linear', name='insulin_output')
    ])

    return model

# Create the DEEP GRU model
model = create_insulin_gru_deep(sequence_length=6, n_features=n_features)

print("USING DEEP GRU MODEL!")
print("Model Architecture:")
model.summary()

# Compile the model
model.compile(
    optimizer='adam',
    loss='mse',
    metrics=['mae', 'mape']
)

# Define callbacks
save_path = '/content/drive/MyDrive/insulin_model/best_insulin_gru_model.keras'

callbacks = [
    tf.keras.callbacks.EarlyStopping(
        monitor='val_loss',
        patience=5,
        min_delta=1e-4,
        restore_best_weights=True
    ),
    tf.keras.callbacks.ReduceLROnPlateau(
        monitor='val_loss',
        factor=0.5,
        patience=3,
        min_lr=1e-6,
        verbose=1
    ),
    tf.keras.callbacks.ModelCheckpoint(
        filepath=save_path,
        monitor='val_loss',
        save_best_only=True,
        verbose=1
    )
]

# Train the model
history = model.fit(
    X_train, y_train,
    batch_size=32,
    epochs=15,
    validation_data=(X_val, y_val),
    callbacks=callbacks,
    verbose=1
)

# Save the model
model.save('insulin_gru_converged.h5')
print("Model saved as 'insulin_gru_converged.h5'")

# Save weights
model.save_weights('insulin_gru_weights.weights.h5')
print("Weights saved as 'insulin_gru_weights.weights.h5'")

# Make predictions
y_pred = model.predict(X_test)

# Plot results
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 4))

# Scatter: actual vs predicted
plt.subplot(1, 2, 1)
plt.scatter(y_test, y_pred, alpha=0.5)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', lw=2)
plt.xlabel('Actual Insulin Need')
plt.ylabel('Predicted Insulin Need')
plt.title('Actual vs Predicted - GRU Model')

# Line plot: first 100 points
plt.subplot(1, 2, 2)
plt.plot(y_test[:100], label='Actual', alpha=0.7)
plt.plot(y_pred[:100], label='Predicted', alpha=0.7)
plt.xlabel('Time Steps')
plt.ylabel('Insulin Need')
plt.title('Sample Predictions - GRU Model')
plt.legend()

plt.tight_layout()
plt.show()

# Calculate metrics
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"\nGRU Model Performance Metrics:")
print(f"MSE: {mse:.4f}")
print(f"MAE: {mae:.4f}")
print(f"R: {r2:.4f}")
```

```python
# =============================================================================
# FEATURE IMPORTANCE ANALYSIS - FIXED VERSION
# =============================================================================

print(" ANALYZING FEATURE IMPORTANCE...")

# Method 1: Manual Permutation Importance (Fixed)
def manual_permutation_importance(model, X, y, n_repeats=5):
    """Manual permutation importance since sklearn doesn't work with Keras models"""
    baseline_score = model.evaluate(X, y, verbose=0)[1]  # Use MAE as metric
    feature_importance = np.zeros(X.shape[2])

    for feature_idx in range(X.shape[2]):
        feature_scores = []

        for _ in range(n_repeats):
            # Create perturbed copy
            X_perturbed = X.copy()

            # Shuffle this feature across samples
            np.random.shuffle(X_perturbed[:, :, feature_idx])

            # Calculate score with perturbed feature
            perturbed_score = model.evaluate(X_perturbed, y, verbose=0)[1]
            feature_scores.append(perturbed_score)

        # Importance = how much worse performance gets when feature is shuffled
        feature_importance[feature_idx] = baseline_score - np.mean(feature_scores)

    return feature_importance

# Calculate manual permutation importance
print("Calculating permutation importance (this may take a minute)...")
manual_importance = manual_permutation_importance(model, X_test[:1000], y_test[:1000], n_repeats=3)

# Get feature names
feature_names = [
    'glucose', 'calories', 'heart_rate', 'steps',
    'basal_rate', 'bolus_volume_delivered', 'carb_input',
    'glucose_change', 'glucose_rolling_mean',
    'calories_cumulative', 'steps_cumulative',
    'sin_hour', 'cos_hour'
]

# Plot feature importance
plt.figure(figsize=(15, 10))

# Sort features by importance
sorted_idx = manual_importance.argsort()[::-1]
sorted_features = [feature_names[i] for i in sorted_idx]
sorted_importance = manual_importance[sorted_idx]

# Create bar plot
plt.subplot(2, 2, 1)
bars = plt.barh(sorted_features, sorted_importance)
plt.xlabel('Permutation Importance (MAE decrease when shuffled)')
plt.title('Feature Importance for Insulin Prediction')
plt.gca().invert_yaxis()

# Add value labels on bars
for bar, value in zip(bars, sorted_importance):
    if value > 0:
        plt.text(bar.get_width() + 0.001, bar.get_y() + bar.get_height()/2,
                 f'{value:.4f}', ha='left', va='center')

# Method 2: Correlation with target (Fixed)
plt.subplot(2, 2, 2)
# Use the last time step of each sequence (most recent values)
last_timestep_features = X_test[:, -1, :]  # Last timestep of each sequence

correlations = []
for i in range(n_features):
    corr = np.corrcoef(last_timestep_features[:, i], y_test)[0, 1]
    correlations.append(corr)

# Sort by absolute correlation
corr_sorted_idx = np.argsort(np.abs(correlations))[::-1]
corr_sorted_features = [feature_names[i] for i in corr_sorted_idx]
corr_sorted_values = [correlations[i] for i in corr_sorted_idx]

bars = plt.barh(corr_sorted_features, corr_sorted_values, color='orange')
plt.xlabel('Correlation with Insulin Need')
plt.title('Feature-Target Correlation (Last Timestep)')
plt.gca().invert_yaxis()

for bar, value in zip(bars, corr_sorted_values):
    plt.text(bar.get_width() + 0.001 if value >= 0 else bar.get_width() - 0.01,
             bar.get_y() + bar.get_height()/2,
             f'{value:.3f}', ha='left' if value >= 0 else 'right', va='center')

# Method 3: Feature std dev analysis (what varies most)
plt.subplot(2, 2, 3)
feature_std = X_test.std(axis=(0, 1))  # Std across all sequences and timesteps

std_sorted_idx = feature_std.argsort()[::-1]
std_sorted_features = [feature_names[i] for i in std_sorted_idx]
std_sorted_values = feature_std[std_sorted_idx]

bars = plt.barh(std_sorted_features, std_sorted_values, color='green')
plt.xlabel('Standard Deviation')
plt.title('Feature Variability')
plt.gca().invert_yaxis()

for bar, value in zip(bars, std_sorted_values):
    plt.text(bar.get_width() + 0.001, bar.get_y() + bar.get_height()/2,
             f'{value:.3f}', ha='left', va='center')

# Method 4: Your original reasoning validation
plt.subplot(2, 2, 4)
# Categorize features by your original intent
categories = {
    'Glucose Dynamics': ['glucose', 'glucose_change', 'glucose_rolling_mean'],
    'Nutrition': ['calories', 'carb_input', 'calories_cumulative'],
    'Insulin Delivery': ['basal_rate', 'bolus_volume_delivered'],
    'Activity': ['heart_rate', 'steps', 'steps_cumulative'],
    'Circadian': ['sin_hour', 'cos_hour']
}

category_scores = {}
for category, features in categories.items():
    indices = [feature_names.index(f) for f in features if f in feature_names]
    if indices:
        # Use average of absolute correlation as score
        category_corrs = [abs(correlations[i]) for i in indices]
        category_scores[category] = np.mean(category_corrs)

plt.barh(list(category_scores.keys()), list(category_scores.values()), color='purple')
plt.xlabel('Average |Correlation|')
plt.title('Feature Category Effectiveness')
plt.tight_layout()
plt.show()

# Print insights
print("\n FEATURE INSIGHTS:")
print("="*50)
print("Top 5 Most Important Features (Permutation):")
for i, (feature, importance) in enumerate(zip(sorted_features[:5], sorted_importance[:5])):
    print(f"  {i+1}. {feature}: {importance:.4f}")

print(f"\nTop 5 Most Correlated Features:")
for i, (feature, corr) in enumerate(zip(corr_sorted_features[:5], corr_sorted_values[:5])):
    print(f"  {i+1}. {feature}: {corr:.4f}")

print(f"\n YOUR FEATURE SELECTION VALIDATION:")
glucose_in_top = any('glucose' in f for f in sorted_features[:3])
insulin_in_top = any(f in ['bolus_volume_delivered', 'basal_rate'] for f in sorted_features[:5])
nutrition_in_top = any(f in ['carb_input', 'calories'] for f in sorted_features[:5])

print(f"  - Glucose-related features in top 3: {' YES' if glucose_in_top else ' NO'}")
print(f"  - Insulin-related features in top 5: {' YES' if insulin_in_top else ' NO'}")
print(f"  - Nutrition features in top 5: {' YES' if nutrition_in_top else ' NO'}")

print(f"\n RECOMMENDATIONS:")
if glucose_in_top and insulin_in_top:
    print("   Your glucose and insulin features are well-chosen!")
else:
    print("    Consider focusing more on glucose and insulin features")

if not nutrition_in_top:
    print("   Nutrition features might be less important than expected")

# Check if engineered features are useful
engineered_features = ['glucose_change', 'glucose_rolling_mean', 'calories_cumulative', 'steps_cumulative']
engineered_importance = [manual_importance[feature_names.index(f)] for f in engineered_features if f in feature_names]
if any(imp > 0 for imp in engineered_importance):
    print("   Your engineered features are adding value!")
else:
    print("   Your engineered features might not be very impactful")
```

```python
# =============================================================================
# ACCURATE FEATURE IMPORTANCE ANALYSIS - FIXED
# =============================================================================

print(" CALCULATING ACCURATE FEATURE IMPORTANCE...")

# Method 1: GRU Gate Analysis (Most Accurate for RNNs)
def analyze_gru_gates(model, feature_names):
    """Analyze what the GRU is actually paying attention to"""

    # Get the first GRU layer's input weights
    for layer in model.layers:
        if isinstance(layer, GRU):
            input_weights = layer.get_weights()[0]  # Shape: (n_features, units*3)
            break
    else:
        # If no GRU layer found, return zeros
        return np.zeros(len(feature_names))

    # Sum absolute weights across all gates and units
    feature_importance = np.sum(np.abs(input_weights), axis=1)

    return feature_importance

# Calculate GRU-based importance
gru_importance = analyze_gru_gates(model, feature_names)

# Method 2: Single Feature Correlation
def single_feature_correlation(X_test, y_test, feature_names):
    """Test correlation of each feature with target"""
    single_feature_scores = []

    for i, feature_name in enumerate(feature_names):
        # Use the last timestep values (most recent)
        feature_values = X_test[:, -1, i]

        # Calculate correlation with target
        correlation = np.corrcoef(feature_values, y_test)[0, 1]
        single_feature_scores.append(abs(correlation))

    return np.array(single_feature_scores)

single_scores = single_feature_correlation(X_test, y_test, feature_names)

# Method 3: Simple Ablation Study
def simple_ablation_importance(model, X_test, y_test, feature_names):
    """How much does correlation drop when we remove temporal patterns?"""
    importance_scores = []

    baseline_score = model.evaluate(X_test, y_test, verbose=0)[1]  # MAE

    for i, feature_name in enumerate(feature_names):
        # For simplicity, use feature variability as importance proxy
        # Features that vary more likely carry more information
        feature_variability = np.std(X_test[:, :, i])
        importance_scores.append(feature_variability)

    return np.array(importance_scores)

ablation_importance = simple_ablation_importance(model, X_test, y_test, feature_names)

# Method 4: Time-aware Importance
def time_aware_importance(X_test, y_test, feature_names):
    """Importance based on correlation at each timestep"""
    time_importance = np.zeros(len(feature_names))

    for i, feature_name in enumerate(feature_names):
        timestep_correlations = []
        for t in range(6):  # For each timestep
            correlation = abs(np.corrcoef(X_test[:, t, i], y_test)[0, 1])
            timestep_correlations.append(correlation)

        # Weight recent timesteps more heavily
        weights = [0.1, 0.15, 0.2, 0.25, 0.3, 1.0]  # Increasing importance
        time_importance[i] = np.average(timestep_correlations, weights=weights)

    return time_importance

time_importance = time_aware_importance(X_test, y_test, feature_names)

# Plot ALL methods for comparison
plt.figure(figsize=(16, 12))

methods = {
    'GRU Gate Weights': gru_importance,
    'Feature-Target Correlation': single_scores,
    'Feature Variability': ablation_importance,
    'Time-Weighted Importance': time_importance
}

for idx, (method_name, importance) in enumerate(methods.items(), 1):
    plt.subplot(2, 2, idx)

    # Normalize importance
    if np.sum(importance) > 0:
        importance_normalized = importance / np.sum(importance)
    else:
        importance_normalized = importance

    # Sort features
    sorted_idx = importance_normalized.argsort()[::-1]
    sorted_features = [feature_names[i] for i in sorted_idx]
    sorted_values = importance_normalized[sorted_idx]

    bars = plt.barh(sorted_features, sorted_values)
    plt.xlabel('Normalized Importance')
    plt.title(f'{method_name}')
    plt.gca().invert_yaxis()

    # Add value labels - FIXED SYNTAX
    for bar, value in zip(bars, sorted_values):
        if value > 0.01:  # Only label significant values
            plt.text(bar.get_width() + 0.001, bar.get_y() + bar.get_height()/2,
                    f'{value:.3f}', ha='left', va='center')

plt.tight_layout()
plt.show()

# Clinical Sense Check
print("\n" + "="*60)
print(" CLINICAL SENSE VALIDATION")
print("="*60)

# What SHOULD be important based on medical knowledge
clinically_important = {
    'CRITICAL': ['glucose', 'carb_input', 'bolus_volume_delivered', 'basal_rate'],
    'IMPORTANT': ['glucose_change', 'glucose_rolling_mean', 'heart_rate'],
    'CONTEXTUAL': ['sin_hour', 'cos_hour', 'calories', 'steps'],
    'DERIVED': ['calories_cumulative', 'steps_cumulative']
}

# Check if results make clinical sense
print("Expected Clinical Importance vs Actual Results:")
print("-" * 50)

# Calculate average importance across all methods
avg_importance = np.zeros(len(feature_names))
for method_importance in methods.values():
    if np.sum(method_importance) > 0:
        normalized = method_importance / np.sum(method_importance)
        avg_importance += normalized
avg_importance /= len(methods)

# Sort by average importance
final_sorted_idx = avg_importance.argsort()[::-1]
final_sorted_features = [feature_names[i] for i in final_sorted_idx]
final_sorted_importance = avg_importance[final_sorted_idx]

print("\n FINAL FEATURE IMPORTANCE RANKING:")
print("-" * 40)
for i, (feature, importance) in enumerate(zip(final_sorted_features, final_sorted_importance), 1):
    category = "???"
    for cat, features in clinically_important.items():
        if feature in features:
            category = cat
            break

    print(f"{i:2d}. {feature:20} {importance:.4f} [{category}]")

# Validate clinical sense
critical_features_rank = []
for feature in clinically_important['CRITICAL']:
    if feature in final_sorted_features:
        rank = final_sorted_features.index(feature) + 1
        critical_features_rank.append(rank)
        print(f"   {feature}: ranked #{rank}")

if all(rank <= 8 for rank in critical_features_rank):  # Top 8 should include critical features
    print("\n EXCELLENT: Feature importance makes clinical sense!")
    print("   Critical diabetes features are properly recognized")
elif all(rank <= 12 for rank in critical_features_rank):
    print("\n  ACCEPTABLE: Most clinical features are recognized")
    print("   But some important features might be underweighted")
else:
    print("\n PROBLEMATIC: Clinical features not properly prioritized")
    print("   The model might be learning spurious patterns")

print(f"\n KEY INSIGHTS:")
# Check if glucose is top 3
glucose_rank = final_sorted_features.index('glucose') + 1 if 'glucose' in final_sorted_features else 999
carbs_rank = final_sorted_features.index('carb_input') + 1 if 'carb_input' in final_sorted_features else 999
insulin_rank = final_sorted_features.index('bolus_volume_delivered') + 1 if 'bolus_volume_delivered' in final_sorted_features else 999

print(f"   - Glucose rank: #{glucose_rank} {'' if glucose_rank <= 3 else ''}")
print(f"   - Carbs rank: #{carbs_rank} {'' if carbs_rank <= 5 else ''}")
print(f"   - Insulin rank: #{insulin_rank} {'' if insulin_rank <= 5 else ''}")

print(f"\n RECOMMENDATION:")
if glucose_rank <= 3 and carbs_rank <= 5 and insulin_rank <= 5:
    print("   Your model is learning clinically meaningful patterns!")
    print("   The feature importance is accurate and makes sense")
else:
    print("   Consider feature engineering to emphasize clinical relationships")
    print("   The model might benefit from more explicit glucose-insulin features")
```
